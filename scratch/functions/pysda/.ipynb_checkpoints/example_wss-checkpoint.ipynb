```python
# let's check out the www.avaiablitlty function
# this let's you download SSURGO availablity
# as pdf, jpg, or shp (zipped) this is default
import wss
wss.availability(form=None, dest=r'C:\temp\test')
```

```python
# next if we want to download
# soil survey areas we need to get
# some basic info so we can get the 
# download urls
allssa = wss.ssurgo()
allssa.head()
```

```python
# that was all of the areas
# most people will never want that
# so we can filter on either a state
# such as 'CA' or search the areanames
state = wss.ssurgo(state='DE')
state.head()
```

```python
# wee see that the filter worked
# because there are only 3 areas
# in Deleware. Now lets try a search filter
# note you can't use state and search together
hamilton = wss.ssurgo(search='hamilton')
hamilton
```

```python
# looks like there are 10
# lets filter our data frame
# to just get a few.
import re
hamilton = hamilton[hamilton['areaname'].str.contains('ohi|flo', flags = re.IGNORECASE)]
hamilton
```

```python
# now we can download the data
# by passing, in this case, the 
# data frame.  this is preferrable
# note the frame paramater can be any
# iterable of valid URLs
# additionally we will unpack these
# and rename to convention 
wss.download(frame=hamilton, dest=r'C:\temp\test', unpack=True, rename=True)

```

```python
# we can see that we have
# received what we have asked for
import os
os.listdir(r'C:\temp\test')
```
